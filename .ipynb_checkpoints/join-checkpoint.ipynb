# Analisis y Mezcla de los datos

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
#Read Data
DATA_NAME = 'resultados/resultados_parte{}.csv'
data_parts = [9,10,11,12]
csv_objects = []

for part in data_parts:
    csv_objects.append(pd.read_csv(DATA_NAME.format(part)))
    
    
```

```python
#Join Data

data = csv_objects[1]

for csv_object in csv_objects[1:]:
    data = pd.merge(data, csv_object, how='outer')
    
```

```python
# See data.
data
```

##### Filtramos por Descripcion no nula. Algo mas?

```python
# Filter empty description 
filteredData = data[data.description.notnull()]
#filteredData = data[filteredData.title.notnull() ]
```

##### Guardamos los datos

```python
# Save Data...
filteredData.to_csv(os.getcwd() +'/resultados/all_data.tsv', index = False, sep = "\t" )
```
